```python
import mplstereonet
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
m200 = pd.read_excel('stereo.xlsx',sheet_name=0,index_col='Unnamed: 0')
```

```python
m0 = pd.read_excel('stereo.xlsx',sheet_name=1,index_col='Unnamed: 0')
```

```python
m200_az = m200['Azimuth in seismic'].tolist()
m200_dip = m200['Dip in seismic'].tolist()
m0_az = m0['Azimuth in seismic'].tolist()
m0_dip =  m0['Dip in seismic'].tolist()
```

```python
strikes = m200_az
dips = m200_dip
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(121, projection='stereonet')
ax.pole(strikes, dips, c='k', label='Pole of the Planes')
ax.density_contourf(strikes, dips, measurement='poles', cmap='Reds')
ax.set_title('Densidade de Estruturas - 200m de Extenso - '+str(len(m200))+' amostras', y=1.10, fontsize=15)
ax.grid()
```
